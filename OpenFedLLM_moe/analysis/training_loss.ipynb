```python
import numpy as np
import matplotlib.pyplot as plt

# 加载训练损失数据
training_loss = np.load('/home/yanlutao/research/FED/OpenFedLLM_moe/output/moe/training_loss.npy')
num_clients, num_rounds = training_loss.shape

# 选择几个客户端进行绘制
selected_clients = [1, 4, 12]  # 选择第2、5、13个客户端

# 创建一个掩码数组，忽略值为 -1 的位置
masked_loss = np.ma.masked_equal(training_loss, -1)

# 绘制选择的客户端的训练损失散点图
plt.figure(figsize=(10, 6))
for client in selected_clients:
    valid_rounds = np.where(masked_loss[client] != -1)[0]
    valid_losses = masked_loss[client][valid_rounds]
    print(f"Client {client+1} valid rounds: {valid_rounds}")
    print(f"Client {client+1} valid losses: {valid_losses}")
    plt.scatter(valid_rounds, valid_losses, label=f'Client {client+1}')

plt.xlabel('Round')
plt.ylabel('Training Loss')
plt.title('Training Loss over Rounds for Selected Clients')
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# 加载训练损失数据
# training_loss = np.load('/home/yanlutao/research/FED/OpenFedLLM_moe/output/moe/training_loss_moe1.npy')
training_loss = np.load('/home/yanlutao/research/output/training_loss_balance.npy')

num_clients, num_rounds = training_loss.shape

# 创建一个掩码数组，忽略值为 -1 的位置
masked_loss = np.ma.masked_equal(training_loss, -1)

# 绘制所有客户端的训练损失散点图
plt.figure(figsize=(15, 8))
for client in range(num_clients):
    valid_rounds = np.where(masked_loss[client] != -1)[0]
    valid_losses = masked_loss[client][valid_rounds]
    plt.scatter(valid_rounds, valid_losses, label=f'Client {client+1}', alpha=0.5)  # 设置透明度alpha

plt.xlabel('Round')
plt.ylabel('Training Loss')
plt.title('Training Loss over Rounds for All Clients')
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1))  # 将图例放在右上角，避免覆盖数据点
plt.grid(True)
# output_path = '/home/yanlutao/research/FED/OpenFedLLM_moe/output/moe/training_loss_plot.png'
# plt.savefig(output_path)
plt.show()
```

```python
# 创建一个掩码数组，忽略值为 -1 的位置
masked_loss = np.ma.masked_equal(training_loss, -1)

# 计算每轮的平均训练损失，忽略 -1 值
average_loss_per_round = np.ma.mean(masked_loss, axis=0)

# 打印前两轮的平均损失
print("Average loss for the first two rounds:")
for round in range(2):
    print(f"Round {round + 1}: {average_loss_per_round[round]:.4f}")

# 绘制平均训练损失曲线
plt.figure(figsize=(10, 6))
plt.plot(range(training_loss.shape[1]), average_loss_per_round, label='Average Training Loss')
plt.xlabel('Round')
plt.ylabel('Average Training Loss')
plt.title('Average Training Loss over Rounds')
plt.legend()
plt.show()
```
