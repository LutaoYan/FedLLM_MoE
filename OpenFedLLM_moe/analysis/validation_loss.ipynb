```python
import numpy as np
import matplotlib.pyplot as plt

# Load the validation loss data from the .npy file
validation_loss_path = '/home/yanlutao/research/output/validation_loss_balance.npy'
validation_loss = np.load(validation_loss_path)

# Get the first and last epoch loss
first_epoch_loss = validation_loss[0]
last_epoch_loss = validation_loss[35]

# Print the first and last epoch loss
print(f"First epoch loss: {first_epoch_loss}")
print(f"Last epoch loss: {last_epoch_loss}")

# Plot the validation loss curve
plt.figure(figsize=(10, 6))
plt.plot(validation_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Validation Loss Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load the validation loss data from the uploaded file
validation_loss_path = '/home/yanlutao/research/FED/OpenFedLLM_moe/output/moe/validation_loss_6.npy'
validation_loss = np.load(validation_loss_path)

# Basic statistics
mean_loss = np.mean(validation_loss)
std_loss = np.std(validation_loss)
min_loss = np.min(validation_loss)
max_loss = np.max(validation_loss)
first_epoch_loss = validation_loss[0]
thirtieth_epoch_loss = validation_loss[29] if len(validation_loss) > 29 else None

# Create a DataFrame for displaying statistics
loss_stats = pd.DataFrame({
    'Mean Loss': [mean_loss],
    'Standard Deviation': [std_loss],
    'Minimum Loss': [min_loss],
    'Maximum Loss': [max_loss],
    'First Epoch Loss': [first_epoch_loss],
    'Thirtieth Epoch Loss': [thirtieth_epoch_loss]
})

print("Validation Loss Statistics:")
print(loss_stats)

# Plot the validation loss curve
plt.figure(figsize=(10, 6))
plt.plot(validation_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Validation Loss Curve')
plt.legend()
plt.grid(True)
plt.show()
```
