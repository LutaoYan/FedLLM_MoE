```python
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 读取 JSON 文件
file_path = '/home/yanlutao/research/output/eval_metrics_obalance.json'
with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# 指定要处理的轮次和层次
target_round = 45  # 第10轮，索引从0开始，所以是9
target_layers = [0, 2, 4, 7]  # 第1, 3, 5, 8个字典，索引从0开始，所以是0, 2, 4, 7

# 提取目标轮次的 aux_stat
aux_stat = data[target_round]['aux_stat']

# 提取指定层次的 T_distribution
t_distributions = [aux_stat[layer]['T_distribution'] for layer in target_layers]

# 设置热力图的尺寸
plt.figure(figsize=(10, 20))

# 生成热力图
for i, t_distribution in enumerate(t_distributions):
    plt.subplot(len(t_distributions), 1, i+1)
    sns.heatmap(np.array(t_distribution).reshape(1, -1), annot=True, fmt="d", cmap="YlGnBu", cbar=False, square=True)
    plt.title(f'Layer {target_layers[i]+1} Token Distribution')
    plt.xlabel('Experts')
    plt.ylabel('')

# 调整子图间距
plt.subplots_adjust(hspace=-0.7)

plt.show()
```
